# Evaluation of the Test Subjects

A total of 28 individuals participated in the evaluation process as test subjects for the comprehensive assessment battery. Two subjects did not complete the questionnaire.


## Libraries

Import the necessary libraries for the evaluation of test subjects.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Data
Read CSV file and create DataFrame.

```python
df = pd.read_csv('data/test_subjects.csv')  
print(df)
```

## Evaluation

### Gender

A pie chart to show the distribution of gender.

```python
plt.figure(figsize=(8.27, 5))
df['Geschlecht'] = df['Geschlecht'].replace({'M�nnlich': 'Male', 'Weiblich': 'Female'})

gender_counts = df['Geschlecht'].value_counts()

colors = ['#9B9B9B', '#c00000']
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12})
plt.title('Gender Distribution', fontsize=14)
plt.ylabel('')

plt.savefig('figures/gender_distribution.pdf', format='pdf')

plt.show()
```

```python
print("Male:", gender_counts['Male'])
print("Female:", gender_counts['Female'])
```

### Age

```python
plt.figure(figsize=(8.27, 6))
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']
df['Altersgruppe'] = pd.cut(df['Alter'], bins=bins, labels=labels, right=False)

age_group_counts = df['Altersgruppe'].value_counts(sort=False)
age_group_counts.plot(kind='bar', color='#9B9B9B')

plt.title('Age Distribution', fontsize=14)  
plt.xlabel('Age Group', fontsize=12)  
plt.ylabel('Number', fontsize=12)  
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12) 

plt.savefig('figures/age_distribution.pdf', format='pdf')

plt.show()
```

```python
youngest = df['Alter'].min()
oldest = df['Alter'].max()

print(f"Youngest test person: {youngest} years old")
print(f"Oldest test person: {oldest} years old")
```

### Previous Experience with RoboTrainer 

```python
plt.figure(figsize=(8.27, 6))

df['Experience'] = df['Wie viel Vorerfahrung haben Sie mit RoboTrainer?'].replace({
    'Noch nie verwendet': 'never used',
    'Einmal kurz verwendet': 'used once briefly'
})

experience_counts = df['Experience'].value_counts()
experience_counts.plot(kind='bar', color='#9B9B9B')

plt.title('Experience with RoboTrainer', fontsize=14)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)
plt.xticks(rotation=0, fontsize=12) 
plt.yticks(fontsize=11)

plt.savefig('figures/experience_distribution.pdf', format='pdf')

plt.show()
```

```python
print("Never used RoboTrainer:", experience_counts['never used'])
print("Used RoboTrainer briefly:", experience_counts['used once briefly'])
```
